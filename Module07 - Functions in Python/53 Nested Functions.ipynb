# Nested Functions

```python
def f():
    def g():
        print("Inside function g")
    print("Inside function f")

f()
```

```python
def f():
    def g():
        print("Inside function g")
    print("Inside function f")

f()
g()
```

```python
def f():
    def g():
        print("Inside function g")

    g()
    print("Inside function f")

f()
```

```python
def f():
    def g():
        print("Inside function g")
        f()

    g()
    print("Inside function f")

f()
```
