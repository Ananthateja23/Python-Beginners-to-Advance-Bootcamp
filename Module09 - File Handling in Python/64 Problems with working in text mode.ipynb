```python
# working with binary files
with open("Images/Image03.png", "r") as f:
    print(f.read())
```

```python
with open("Images/Image03.png", "rb") as f:
    print(f.read())
```

```python
with open("Images/Image03.png", "rb") as f:
    with open("files/Image03_copy.png", "wb") as wf:
        data = f.read()
        wf.write(data)
```
